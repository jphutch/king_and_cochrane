```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import matplotlib

years = [1910,1920,1930,1940,1950,1959]
```

```python
acre = pd.read_csv("../data/Census1959_Acres.csv",header=None)

acre = acre.T

acre.columns = acre.iloc[0,:]

acre = acre.iloc[3:]

acre = acre.apply(lambda x: x.str.replace(",",""),axis=1)

acre = acre.apply(lambda x: pd.to_numeric(x,errors="coerce"))

acre = acre.set_index("year").sort_index()

acre['operators_tenants_noncash'] = acre['operators_tenants_all'] - acre['operators_tenants_cash']
acre['operators_white_tenants_noncash'] = acre['operators_white_tenants_all'] - acre['operators_white_tenants_cash']
acre['operators_nonwhite_tenants_noncash'] = acre['operators_nonwhite_tenants_all'] - acre['operators_nonwhite_tenants_cash']

acre['operators_nonmange'] = acre['operators_all'] - acre['operators_managers']
acre['operators_white_nonmange'] = acre['operators_white_all'] - acre['operators_white_managers']
acre['operators_nonwhite_nonmange'] = acre['operators_nonwhite_all'] - acre['operators_nonwhite_managers']

ops = pd.read_csv("../data/Census1959_Ops.csv",header=None).T

ops.columns =[x.strip() for x in ops.iloc[0,:]]

ops = ops.iloc[1:,:]

ops = ops.apply(lambda x: pd.to_numeric(x,errors="coerce"))

ops = ops.set_index("year").sort_index()

ops['operators_tenants_noncash'] = ops['operators_tenants_all'] - ops['operators_tenants_cash']
ops['operators_white_tenants_noncash'] = ops['operators_white_tenants_all'] - ops['operators_white_tenants_cash']
ops['operators_nonwhite_tenants_noncash'] = ops['operators_nonwhite_tenants_all'] - ops['operators_nonwhite_tenants_cash']

ops['operators_nonmange'] = ops['operators_all'] - ops['operators_managers']
ops['operators_white_nonmange'] = ops['operators_white_all'] - ops['operators_white_managers']
ops['operators_nonwhite_nonmange'] = ops['operators_nonwhite_all'] - ops['operators_nonwhite_managers']

south = pd.read_csv("../data/Census1959_South.csv").T.reset_index()
south.columns = south.loc[0,:]
south = south.iloc[1:,:]

south = south.apply(lambda x: pd.to_numeric(x.str.replace(",",""),errors="coerce"))

south = south.set_index(["year"])

ops['operators_nonsouth_nonwhite_tenants'] = ops['operators_nonwhite_tenants_all'] - south.loc[:,"op_south_nonwhite_tenants"]
ops['operators_nonsouth_nonwhite_owners'] = ops['operators_nonwhite_full'] - south.loc[:,"op_south_nonwhite_owners"]
ops['operators_nonsouth_nonwhite_part_owners'] = ops['operators_nonwhite_part'] - south.loc[:,"op_south_nonwhite_part_owners"]

ops['operators_nonsouth_white_tenants'] = ops['operators_white_tenants_all'] - south.loc[:,"op_south_white_tenants"]
ops['operators_nonsouth_white_owners'] = ops['operators_white_full'] - south.loc[:,"op_south_white_owners"]
ops['operators_nonsouth_white_part_owners'] = ops['operators_white_part'] - south.loc[:,"op_south_white_part_owners"]

acre['operators_nonsouth_nonwhite_tenants'] = acre['operators_nonwhite_tenants_all'] - south.loc[:,"acres_south_nonwhite_tenants"]
acre['operators_nonsouth_nonwhite_owners'] = acre['operators_nonwhite_full'] - south.loc[:,"acres_south_nonwhite_owners"]
acre['operators_nonsouth_nonwhite_part_owners'] = acre['operators_nonwhite_part'] - south.loc[:,"acres_south_nonwhite_part_owners"]

acre['operators_nonsouth_white_tenants'] = acre['operators_white_tenants_all'] - south.loc[:,"acres_south_white_tenants"]
acre['operators_nonsouth_white_owners'] = acre['operators_white_full'] - south.loc[:,"acres_south_white_owners"]
acre['operators_nonsouth_white_part_owners'] = acre['operators_white_part'] - south.loc[:,"acres_south_white_part_owners"]

Acre_Region = pd.concat([acre,south],axis=1)
Ops_Region = pd.concat([ops,south],axis=1)

ops_pctchngs = Ops_Region.loc[[1920,1959]].pct_change().loc[1959]
acre_pctchngs = Acre_Region.loc[[1920,1959]].pct_change().loc[1959]

owner_vars = ["operators_white_full","operators_white_part","operators_white_tenants_all",
              'operators_nonwhite_full','operators_nonwhite_part',"operators_nonwhite_tenants_all"]

tenant_vars = ['operators_white_tenants_cash',"operators_white_tenants_noncash",'operators_white_tenants_sharecrop',
               'operators_nonwhite_tenants_noncash', 'operators_nonwhite_tenants_cash','operators_nonwhite_tenants_sharecrop']

all_vars = owner_vars + tenant_vars

white_own_tenants_acre = acre[['operators_white_full',
                               "operators_white_part",
                               'operators_white_tenants_all']].sort_index().loc[years]/10**6

nonwhite_own_tenants_acre = acre[['operators_nonwhite_full',
                                  "operators_nonwhite_part",
                                  'operators_nonwhite_tenants_all']].sort_index().loc[years]/10**6

white_own_tenants_ops = ops[['operators_white_full',
                             "operators_white_part",
                             'operators_white_tenants_all']].sort_index().loc[years]/10**6

nonwhite_own_tenants_ops = ops[['operators_nonwhite_full',
                                "operators_nonwhite_part",
                                'operators_nonwhite_tenants_all']].sort_index().loc[years]/10**6

nonwhite_own_tenants = acre[['operators_nonwhite_full',
                             'operators_nonwhite_tenants_all']].sort_index().loc[years]/10**6

white_own_tenants = acre[['operators_white_full',
                          'operators_white_tenants_all']].sort_index().loc[years]/10**6

ops_nonsouth_white = Ops_Region[['operators_nonsouth_white_owners',
                                'operators_nonsouth_white_part_owners',
                                'operators_nonsouth_white_tenants']].sort_index().loc[years]/10**6

ops_nonsouth_nonwhite = Ops_Region[['operators_nonsouth_nonwhite_owners',
                                    'operators_nonsouth_nonwhite_part_owners',
                                    'operators_nonsouth_nonwhite_tenants']].sort_index().loc[years]/10**6

ops_south_white = Ops_Region[['op_south_white_owners',
                              'op_south_white_part_owners',
                              'op_south_white_tenants']].sort_index().loc[years]/10**6

ops_south_nonwhite = Ops_Region[['op_south_nonwhite_owners',
                                 'op_south_nonwhite_part_owners',
                                 'op_south_nonwhite_tenants']].sort_index().loc[years]/10**6

acres_nonsouth_white = Acre_Region[['operators_nonsouth_white_owners',
                                     'operators_nonsouth_white_part_owners',
                                     'operators_nonsouth_white_tenants']].sort_index().loc[years]/10**6

acres_nonsouth_nonwhite = Acre_Region[['operators_nonsouth_nonwhite_owners',
                                       'operators_nonsouth_nonwhite_part_owners',
                                       'operators_nonsouth_nonwhite_tenants']].sort_index().loc[years]/10**6

acres_south_white = Acre_Region[['acres_south_white_owners',
                                 'acres_south_white_part_owners',
                                 'acres_south_white_tenants']].sort_index().loc[years]/10**6

acres_south_nonwhite =  Acre_Region[['acres_south_nonwhite_owners',
                                     'acres_south_nonwhite_part_owners',
                                     'acres_south_nonwhite_tenants']].sort_index().loc[years]/10**6
```

### Table 1 

```python
ops_change =  \
pd.concat([(white_own_tenants_ops.loc[1959]-white_own_tenants_ops.loc[1920])/white_own_tenants_ops.loc[1920]
,(nonwhite_own_tenants_ops.loc[1959]-nonwhite_own_tenants_ops.loc[1920])/nonwhite_own_tenants_ops.loc[1920]],axis=0) *100

white_ops = (pd.concat([white_own_tenants_ops.loc[1920],white_own_tenants_ops.loc[1959]],axis=1)*10**3).round(2)
nonwhite_ops = (pd.concat([nonwhite_own_tenants_ops.loc[1920],nonwhite_own_tenants_ops.loc[1959]],axis=1)*10**3).round(2)

nonwhite_pct_ops = 100*nonwhite_ops.values/(white_ops.values + nonwhite_ops.values)

nonwhite_pct_ops = \
pd.DataFrame(nonwhite_pct_ops,
             columns = [1920,1959],
             index=["ops_nonwhite_pct_full",
                    "ops_nonwhite_pct_part",
                    "ops_nonwhite_pct_tenants"])

ops_totals = \
pd.concat([pd.DataFrame(white_ops.sum()),\
           pd.DataFrame(nonwhite_ops.sum()),
           pd.DataFrame(100*nonwhite_ops.sum() /(white_ops.sum() +nonwhite_ops.sum()))],axis=1).T

ops_totals.index = ["ops_white_total","ops_nonwhite_total","ops_pct_total"]

ops_table = pd.concat([pd.concat([white_ops,nonwhite_ops,nonwhite_pct_ops,ops_totals])],axis=1)

ops_table.columns = ["1920","1959"]

ops_table = \
ops_table.loc[['operators_white_full',"operators_nonwhite_full","ops_nonwhite_pct_full",
           'operators_white_part',"operators_nonwhite_part","ops_nonwhite_pct_part",
           "operators_white_tenants_all","operators_nonwhite_tenants_all","ops_nonwhite_pct_tenants",
           "ops_white_total","ops_nonwhite_total","ops_pct_total"]]

ops_table['type'] = ["Owner"]*3 + ["Part Owner"]*3 +["Tenant"]*3 +["All"]*3
ops_table['race'] = ["White","Non-white","% Non-white"]*4

ops_table['Change'] = ops_table['1959'] - ops_table['1920']
ops_table["% Change"] = 100* (ops_table['1959'] - ops_table['1920'])/ops_table["1920"]

ops_table = ops_table.set_index(["type","race"])



acres_change =  \
pd.concat([(white_own_tenants_acre.loc[1959]-white_own_tenants_acre.loc[1920])/white_own_tenants_acre.loc[1920]
,(nonwhite_own_tenants_acre.loc[1959]-nonwhite_own_tenants_acre.loc[1920])/nonwhite_own_tenants_acre.loc[1920]],axis=0) *100

white_acres = (pd.concat([white_own_tenants_acre.loc[1920],white_own_tenants_acre.loc[1959]],axis=1)*10**1).round(2)
nonwhite_acres = (pd.concat([nonwhite_own_tenants_acre.loc[1920],nonwhite_own_tenants_acre.loc[1959]],axis=1)*10**1).round(2)

nonwhite_pct_acres = 100*nonwhite_acres.values/(white_acres.values + nonwhite_acres.values)

nonwhite_pct_acres = \
pd.DataFrame(nonwhite_pct_acres,
             columns = [1920,1959],
             index=["acres_nonwhite_pct_full",
                    "acres_nonwhite_pct_part",
                    "acres_nonwhite_pct_tenants"])

acres_totals = \
pd.concat([pd.DataFrame(white_acres.sum()),\
           pd.DataFrame(nonwhite_acres.sum()),
           pd.DataFrame(100*nonwhite_acres.sum() /(white_acres.sum() +nonwhite_acres.sum()))],axis=1).T

acres_totals.index = ["acres_white_total","acres_nonwhite_total","acres_pct_total"]

acre_table = pd.concat([pd.concat([white_acres,nonwhite_acres,nonwhite_pct_acres,acres_totals])],axis=1)

acre_table.columns = ["1920","1959"]

acre_table = \
acre_table.loc[['operators_white_full',"operators_nonwhite_full","acres_nonwhite_pct_full",
           'operators_white_part',"operators_nonwhite_part","acres_nonwhite_pct_part",
           "operators_white_tenants_all","operators_nonwhite_tenants_all","acres_nonwhite_pct_tenants",
           "acres_white_total","acres_nonwhite_total","acres_pct_total"]]

acre_table['type'] = ["Owner"]*3 + ["Part Owner"]*3 +["Tenant"]*3 +["All"]*3
acre_table['race'] = ["White","Non-white","% Non-white"]*4

acre_table['Change'] = acre_table['1959'] - acre_table['1920']
acre_table["% Change"] = 100* (acre_table['1959'] - acre_table['1920'])/acre_table["1920"]

acre_table = acre_table.set_index(["type","race"])

acre_table.index.names = ["Acres (millions)",""]
ops_table.index.names = ["Operators",""]


ops_tenure_total = \
pd.concat([ops_table.reset_index()[ops_table.reset_index()[""]!="% Non-white"].groupby(["Operators"])['1920'].sum(),
           ops_table.reset_index()[ops_table.reset_index()[""]!="% Non-white"].groupby(["Operators"])['1959'].sum()],axis=1)

ops_tenure_total['Change'] = ops_tenure_total.diff(axis=1)['1959']
ops_tenure_total['% Change'] = 100*ops_tenure_total.pct_change(axis=1)['1959']

ops_tenure_total[''] = ['Total']*4

ops_tenure_total = ops_tenure_total.reset_index().set_index(["Operators",""])

acre_tenure_total = \
pd.concat([acre_table.reset_index()[acre_table.reset_index()[""]!="% Non-white"].groupby(["Acres (millions)"])['1920'].sum(),
           acre_table.reset_index()[acre_table.reset_index()[""]!="% Non-white"].groupby(["Acres (millions)"])['1959'].sum()],axis=1)

acre_tenure_total['Change'] = acre_tenure_total.diff(axis=1)['1959']
acre_tenure_total['% Change'] = 100*acre_tenure_total.pct_change(axis=1)['1959']

acre_tenure_total[''] = ['Total']*4

acre_tenure_total = acre_tenure_total.reset_index().set_index(["Acres (millions)",""])

acre_table = pd.concat([acre_table,acre_tenure_total])
ops_table = pd.concat([ops_table,ops_tenure_total])
```
